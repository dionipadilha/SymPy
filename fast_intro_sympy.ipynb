# Fast Intro SymPy

SymPy is a powerful tool for symbolic computation, especially for tasks involving algebraic manipulations and solving equations in a symbolic form.

## Symbolic Environment

 SymPy allows you to work in a symbolic environment where variables can represent abstract symbols, rather than specific numerical values.

```python
# Symbolic Environment
from sympy import *
init_printing()
```

```python
# Symbolic Variables
x = Symbol('x')
y = S('y')
a, b, c = symbols('a, b, c')
n = Symbol('n', integer=True)
```

## Rationals

SymPy can handle rational numbers symbolically and provide accurate fractional approximations.

```python
# Numeric Rational
8/4 + 3/7
```

```python
# Symbolic Rational
Rational(8, 4) + Rational(3, 7)
```

```python
# Fractional Approximation
nsimplify(8/4 + 3/7)
```

## Numerical Evaluation

SymPy can evaluate symbolic expressions numerically, including constants like pi.

```python
# Symbolic Constants
pi
```

```python
# Numeric Constants
N(pi)
```

```python
# Numeric Approximation
N(pi, 3)
```

## Expressions

SymPy can manipulate and simplify symbolic expressions in SymPy, making it ideal for algebraic computations.

```python
# Symbolic Expressions
n*(a + b*c - x/y)
```

```python
# Expression Manipulation
expr = (a+b)**2
expr.expand()
```

## Equations

SymPy can represent and solve symbolic equations, both algebraic and transcendental.

```python
# Symbolic Equations
a, b, c, x = symbols('a, b, c, x')
quadratic_eq = Eq(a*x**2 + b*x + c, 0)
quadratic_eq
```

```python
# Symbolic Solving
quadratic_sol = solve(quadratic_eq, x)
quadratic_sol
```

```python
# Substitution
coefficients = {a:1, b:2, c:-3}
solve(quadratic_eq.subs(coefficients), x)
```

```python
# Simplification
expr = (x**2 + x)/(x*sin(y)**2 + x*cos(y)**2)
Eq(expr, simplify(expr))
```

## Systems of Equations

 SymPy can solve systems of equations using functions like linsolve.

```python
# Solving Systems of Equations
eq_1 = Eq(2*x - y, 2)
eq_2 = Eq(x + 2*y, 21)
system = [eq_1, eq_2]
linsolve(system, [x, y])
```

## Summation and Product

SymPy can perform symbolic summation and product operations using Sum and Product.

```python
# Summation
s = Sum(n, (n, 1, 5))
Eq(s, s.doit())
```

```python
# Product
p = Product(n, (n, 1, 5))
Eq(p, p.doit())
```

## Matrices

 SymPy supports matrix operations, such as transposition and determinant calculation.

```python
# Matrices
M = Matrix([[1, 3], [-2, 3]])
M
```

```python
# Transpose of a Matrix
M.T
```

```python
# Determinant of a Matrix
M.det()
```

```python
# Matrices Basic Operations
M * 6
```

## Functions

 SymPy can work with symbolic functions and evaluate them at specific points.

```python
# Commom Builtins Functions
a**x, sqrt(x), log(x, b), factorial(n)
```

```python
# Trigonometric Functions
sin(pi/n), cos(x), tan(2*x)
```

```python
# Define a Functions
x=symbols('x')
f = Lambda(x, sin(x)/x)
f(x)
```

```python
f_= Derivative(f(x), x)
Eq(f_,f_.doit())
```

```python
plot(f(x), (x, -5, 5));
```

```python
L=Limit(f(x), x, 0)
L
```

```python
L.doit()
```

```python
# Function Evaluation
f(3)
```

## Plotting

SymPy provides basic plotting capabilities for functions, parametric plots, and 3D surface plots.

```python
# Plot Functions
plot(f(x), (x, -5, 5));
```

```python
# Plot Multiple Functions
plot(sin(x), cos(x), (x, -pi, pi));
```

```python
# Parametric Plot
plot_parametric((cos(x), sin(x)), (x, -5, 5));
```

```python
# 3D Surface Plot
from sympy.plotting.plot import plot3d
plot3d(sin(x)*cos(y), (x, -5, 5), (y, -5, 5));
```
