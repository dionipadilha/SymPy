# Rational function

```python
# Symbolic env
from sympy import *
a, b, c, d = symbols('a, b, c, d')
x, y, z, t = symbols('x, y, z, t')
k, m, n = symbols('k, m, n', integer=True)
f, g, h = symbols('f, g, h', cls=Function)
init_printing()
```

```python
# Define the function f(x):
p = Poly(x**3 + 3*x**2)
q = Poly(x**2 + 6*x + 9)
f = p/q
f
```

```python
# Configure the plot style
from seaborn import set_style as set_plot_style
set_plot_style({
  'figure.facecolor': '#333',
  'axes.facecolor': '#333',
  'axes.edgecolor': '.75',
  'axes.labelcolor': '.75',
  'axes.grid': True,
  'grid.color':'.75',
  'grid.linestyle':':',
  'xtick.color': '.75',
  'ytick.color': '.75',
  'text.color': '.75'
})

# Plot the graph of f(x):
plot(
  *[f, x + 3],
  (x, -20, 20),
  xlim=(-20, 20),
  ylim=(-20, 20),
  legend=True,
  aspect_ratio=(1.618, 1),
  axis=True,
);
```

```python
# determine the roots of f(x):
solveset(f, x)
```

```python
# identify the behaviour on the left edge:
left_edge = Limit(f, x, -oo)
Eq(left_edge, left_edge.doit())
```

```python
# identify the behaviour on the right edge:
right_edge = Limit(f, x, oo)
Eq(right_edge, right_edge.doit())
```

```python
# simplify f(x):
Eq(f, cancel(f))
```

```python
# identify any potential vertical asymptotes:
n, d = fraction(cancel(f))
d
```

```python
# determine the vertical asymptotes:
maximum_domain = solve(d, x)[0]
maximum_domain
```

```python
# identify the behaviour of f(x) around maximum_domain
LimitBelow = limit(f, x, maximum_domain, '-')
LimitAbove = limit(f, x, maximum_domain, '+')
LimitBelow, LimitAbove
```
