```python
# symbolic env:
from sympy import *
init_printing()
a, b, c, d = symbols('a, b, c, d')
x, y, z, t = symbols('x, y, z, t')
k, m, n = symbols('k, m, n', integer=True)
f, g, h = symbols('f, g, h', cls=Function)
```

```python
# plot style configuration:
from seaborn import set_style as sns_plotstyle
sns_plotstyle({
  'figure.facecolor': '#333',
  'axes.facecolor': '#333',
  'axes.edgecolor': '.75',
  'axes.labelcolor': '.75',
  'axes.grid': True,
  'grid.color':'.75',
  'grid.linestyle':':',
  'xtick.color': '.75',
  'ytick.color': '.75',
  'text.color': '.75'
})
```

```python
from sympy.plotting.plot import MatplotlibBackend

# function to be displayed
f = Lambda(x, sin(x))
p = plot(f(x), (x,-pi,pi), show=False)

# point to be displayed
x0 = pi/2
y0 = f(x0)

# plot with the commands from the backend
be = MatplotlibBackend(p)
be.process_series()
plt = be.plt
plt.plot([-x0, x0], [-y0, y0],'r.', label="mypoint", ms=10, ls=':')

# update plot to include legend and markers
plt.legend()
plt.show()
```
