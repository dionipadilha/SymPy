```python
# symbolic env:
from sympy import *
init_printing()
```

```python
# expressions with one variable:
x = symbols('x')
f = sin(x)**2
f.diff(x)
```

```python
# partial derivatives:
x, y = symbols('x, y')
f = sin(x)**2 + cos(y)**2
f.diff(x)
```

```python
# differentiate the symbolic expression f with respect to a variable y:
f.diff(y)
```

```python
# second partial and mixed derivatives:
f.diff(y, 2)
```

```python
# find the indefinite integral:
x = symbols('x')
f = sin(x)**2
f.integrate(x).simplify()
```

```python
# indefinite integrals of multivariable expressions:
x, y, n = symbols('x, y, n')
f = x**n + y**n
f.integrate(x)
```

```python
# integrate the expression with respect to y:
f.integrate(y)
```

```python
# integrate the expression with respect to n:
f.integrate(n)
```

```python
# definite integral:
x, y, n = symbols('x, y, n')
f = x**2 + 2*x - 3
f.integrate((x, -3, 1))
```

```python
# solve algebraic equations:
solve(Eq(x**3 - 6*x**2, 6 - 11*x))
```

```python
# if you do not specify the right side of the equation:
solve(x**3 - 6*x**2 + 11*x - 6)
```

```python
# solve algebraic equations with several symbolic variables:
x, y = symbols('x, y')
solve(6*x**2 - 6*x**2*y + x*y**2 - x*y + y**3 - y**2, y)
```

```python
# solve systems of equations:
x, y, z = symbols('x, y, z')
solve([Eq(z , 4*x), Eq(x , y), Eq(z , x**2 + y**2)])
```

```python
# create symbolic expressions:
phi = (1 + sqrt(5))/2;
f = phi**2 - phi - 1
f
```

```python
# simplify symbolic expressions:
f.simplify()
```

```python
# rewrite a polynomial in the standard form:
f = (x**2- 1)*(x**4 + x**3 + x**2 + x + 1)*(x**4 - x**3 + x**2 - x + 1)
f.expand()
```

```python
# shows the polynomial roots
g = x**3 + 6*x**2 + 11*x + 6;
g.factor()
```

```python
# polynomial in the most efficient for numerical evaluations:
h = x**5 + x**4 + x**3 + x**2 + x;
horner(h)
```

```python
# substitute a symbolic variable with a numeric value:
f = 2*x**2 - 3*x + 1;
f.subs(x, S(1)/3)
```

```python
# the subs function does not change the original expression f:
f
```

```python
# substitute in multivariate expressions:
x, y = symbols('x, y')
f = x**2*y + 5*x*sqrt(y)
f.subs({x:3, y:4})
```

```python
# substitute one symbolic variable for another:
f.subs({x:y, y:x})
```

```python
# plot symbolic functions:
x = symbols('x')
f = x**3 - 6*x**2 + 11*x - 6
plot(f, (x,-5,5));
```

```python
# options for plot:
from seaborn import set_style as set_plot_style
set_plot_style('whitegrid')
plot(f, (x,-5,5), label=f, legend=True, aspect_ratio=(120,1));
```

```python
# implicit function plot:
eqn = Eq((x**2 + y**2)**4, (x**2 - y**2)**2)
plot_implicit(eqn, (x, -1, 1), (y, -1, 1));
```

```python
# Plot 3-D parametric lines:
from sympy.plotting import plot3d_parametric_line
x, y, z, t = symbols('x, y, z, t')
x = t**2 * sin(10*t)
y = t**2 * cos(10*t)
z = t
plot3d_parametric_line(x, y, z, (t, -5, 5));
```

```python
# Create a 3-D surface
from sympy.plotting import plot3d
x, y = symbols('x, y')
plot3d(x**2 + y**2, (x, -5, 5), (y, -5, 5));
```
