# Phase Shift

```python
# symbolic env:
from sympy import *
init_printing()
a, b, c, d = symbols('a, b, c, d')
x, y, z, t = symbols('x, y, z, t')
k, m, n = symbols('k, m, n', integer=True)
f, g, h = symbols('f, g, h', cls=Function)
```

```python
# plot style:
from seaborn import set_style as sns_plotstyle
sns_plotstyle({
  'figure.facecolor': '#333',
  'axes.facecolor': '#333',
  'axes.edgecolor': '.75',
  'axes.labelcolor': '.75',
  'axes.grid': True,
  'grid.color':'.75',
  'grid.linestyle':':',
  'xtick.color': '.75',
  'ytick.color': '.75',
  'text.color': '.75'
})
```

```python
# phase shift:
theta = (2*pi/3)
waves = [sin(t + 0*theta), sin(t + 1*theta), sin(t + 2*theta)]
plot(*waves, (t, 0, 2*pi));
```
