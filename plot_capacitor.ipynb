```python
# Symbolic Env:
from sympy import init_printing, symbols, Lambda, exp, plot
init_printing()

# Circuit Parameters:
R = 1e+3 # resistance in ohms
C = 1e-3 # capacitance in farads
Vs = 100 # source in volts

# Capacitor Charge:
t = symbols('t')
τ = R * C
capacitor_charge = Lambda(t, (Vs * (1 - exp(-t / τ))))
capacitor_discharge = Lambda(t, (Vs * exp(-t / τ)))
```

```python
# Plotting
p = plot(
  (capacitor_charge(t), (t, 0, 6*τ)),
  (capacitor_discharge(t), (t, 0, 6*τ)),
  title='Capacitor in RC Circuit',
  xlabel='t',
  ylabel='V(t)',
  legend = True,
  margin=0.15,
  show=False,
)
p[0].label = 'Charging'
p[1].label = 'Discharging'
p.show()
```

```python
# Style
from seaborn import set_style as sns_plotstyle
sns_plotstyle({
  'figure.facecolor': 'none',
  'axes.facecolor': 'none',
  'axes.edgecolor': '.75',
  'axes.labelcolor': '.75',
  'axes.grid': True,
  'grid.color':'.75',
  'grid.linestyle':':',
  'xtick.color': '.75',
  'ytick.color': '.75',
  'text.color': '.75'
})
p.show()
```
