# Plotting

```python
# Configure the symbolic environment
from sympy import symbols, sin, plot, plotting
x, y = symbols('x, y')
```

```python
# Configure the plot style
styles = {
  'figure.facecolor': '#333',
  'axes.facecolor': '#333',
  'axes.edgecolor': '.75',
  'axes.labelcolor': '.75',
  'axes.grid': True,
  'grid.color':'.75',
  'grid.linestyle':':',
  'xtick.color': '.75',
  'ytick.color': '.75',
  'text.color': '.75'
}
from seaborn import set_style as sns_plotstyle
sns_plotstyle(styles)
```

```python
# Plot a 2d function
functions = [sin(x), sin(2*x)]
plot(
  *functions,
  (x, 0, 7),
  legend = True,
  aspect_ratio = (1.618, 1)
);
```

```python
# Plot a 3d function
plotting.plot3d(
  sin(x**2+y**2),
  (x,-1,1),
  (y,-1,1)
);
```
