# Quadratic Equations

```python
# Configure the symbolic environment
from sympy import init_printing, symbols, Function, Eq, solveset, Lambda, plot
init_printing() # latex format
a, b, c, x = symbols('a b c x')

# Configure the plot style
from seaborn import set_style as sns_plotstyle
sns_plotstyle({
  'figure.facecolor': '#fff',
  'axes.facecolor': '#fff',
  'axes.edgecolor': '.5',
  'axes.labelcolor': '.5',
  'axes.grid': True,
  'grid.color':'.5',
  'grid.linestyle':':',
  'xtick.color': '.5',
  'ytick.color': '.5',
  'text.color': '.5'
})
```

```python
# Define symbolic general solution
f = Lambda(x, a*x**2 + b*x + c)
quadratic_equation = Eq(f(x), 0)
roots = solveset(quadratic_equation, x)
roots
```

```python
# Substite the numerical coefficients
coefficients = {a: 1, b: 2, c: -3}
roots.subs(coefficients)
```

```python
# Plot the quadratic function
plot(
  f(x).subs(coefficients),
  (x, -5, 3),
  legend = True,
  aspect_ratio = (1.618, 1) # Golden_Ratio
);
```
