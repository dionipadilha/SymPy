# Quadratic Equations

```python
# Configure the symbolic environment
from sympy import symbols, Function, Eq, solveset, Lambda, plot
a, b, c, x = symbols('a b c x')

# Configure the plot style
from seaborn import set_style as sns_plotstyle
sns_plotstyle({
  'figure.facecolor': '#333',
  'axes.facecolor': '#333',
  'axes.edgecolor': '.75',
  'axes.labelcolor': '.75',
  'axes.grid': True,
  'grid.color':'.75',
  'grid.linestyle':':',
  'xtick.color': '.75',
  'ytick.color': '.75',
  'text.color': '.75'
})
Golden_Ratio = (1.618, 1)
```

```python
# Define symbolic general solution
f = Lambda(x, a*x**2 + b*x + c)
quadratic_equation = Eq(f(x), 0)
general_solution = solveset(quadratic_equation, x)
general_solution
```

```python
# Substite the numerical coefficients
coefficients_values = {a: 1, b: -4, c: 3}
general_solution.subs(coefficients_values)
```

```python
# Plot the quadratic function
plot(
  f(x).subs(coefficients_values),
  (x, -1, 5),
  legend = True,
  aspect_ratio = Golden_Ratio
);
```
