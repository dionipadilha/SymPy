# Quadratic Equations

```python
# Configure the symbolic environment
from sympy import *
init_printing() # latex format
a, b, c, x = symbols('a b c x')

# Configure the plot style
from seaborn import set_style as sns_plotstyle
sns_plotstyle({
  'figure.facecolor': '#fff',
  'axes.facecolor': '#fff',
  'axes.edgecolor': '.5',
  'axes.labelcolor': '.5',
  'axes.grid': True,
  'grid.color':'.5',
  'grid.linestyle':':',
  'xtick.color': '.5',
  'ytick.color': '.5',
  'text.color': '.5'
})
```

```python
# Define symbolic solution
quadratic_expr = a*x**2 + b*x + c
symbolic_roots = roots(quadratic_expr, x)
symbolic_roots
```

```python
# Define numeric solution
n_coefficients = {a: 1, b: 2, c: -3}
n_quadratic_expr = quadratic_expr.subs(n_coefficients)
n_roots = roots(n_quadratic_expr, x)
n_roots
```

```python
# Plot the quadratic function
plot(
  Lambda(x, n_quadratic_expr)(x), # plot_function
  (x, -5, 3),                     # plot_range
  aspect_ratio = (1.618, 1),      # Golden_Ratio
  legend = True
);
```
