# Quadratic Equations

```python
# Configure the symbolic environment
from sympy import *
init_printing() # latex format
a, b, c, x = symbols('a, b, c, x')

# Configure the plot style
from seaborn import set_style as sns_plotstyle
sns_plotstyle({
  'figure.facecolor': '#fff',
  'axes.facecolor': '#fff',
  'axes.edgecolor': '.5',
  'axes.labelcolor': '.5',
  'axes.grid': True,
  'grid.color':'.5',
  'grid.linestyle':':',
  'xtick.color': '.5',
  'ytick.color': '.5',
  'text.color': '.5'
})
```

```python
# Define symbolic objects
quadratic_expr = a*x**2 + b*x + c
quadratic_roots = {x for x,_ in roots(quadratic_expr, x).items()}
quadratic_roots
```

```python
# Define numeric objects
n_coefficients = {a: 1, b: 2, c: -3}
n_quadratic_roots = {x.subs(n_coefficients) for x in quadratic_roots}
n_quadratic_roots
```

```python
# Plot the quadratic function
quadratic_function = Lambda(x, quadratic_expr)(x)
plot_function = quadratic_function.subs(n_coefficients)
plot_range = (x, -5, 3)
Golden_Ratio = (1.618, 1)
plot( plot_function, plot_range, aspect_ratio = Golden_Ratio, legend = True);
```
