```python
# Configure the symbolic environment
from sympy import *
init_printing() # latex format
a, b, c, x = symbols('a, b, c, x', real=True)

# Configure the plot style
from seaborn import set_style as sns_plotstyle
sns_plotstyle({
  'figure.facecolor': '#fff',
  'axes.facecolor': '#fff',
  'axes.edgecolor': '.5',
  'axes.labelcolor': '.5',
  'axes.grid': True,
  'grid.color':'.5',
  'grid.linestyle':':',
  'xtick.color': '.5',
  'ytick.color': '.5',
  'text.color': '.5'
})
```

```python
# Define symbolic objects
quadratic_function = Lambda(x, a*x**2 + b*x + c)(x)
quadratic_eq = Eq(quadratic_function, 0)
sym_roots = solveset(quadratic_eq, x)
sym_roots
```

```python
# Define numeric objects
n_coefficients = {a: 1, b: 2, c: -3}
n_roots = sym_roots.subs(n_coefficients)
n_roots
```

```python
# Plot the quadratic function
plot_function = quadratic_function.subs(n_coefficients)
plot_range = (x, -5, 3)
Golden_Ratio = (1.618, 1)
plot(
  plot_function,
  plot_range,
  aspect_ratio = Golden_Ratio,
  legend = True
);
```
