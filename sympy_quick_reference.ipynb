```python
# Continued Fractions
def list_to_frac(l):
    expr = Integer(0)
    for i in reversed(l[1:]):
        expr += i
        expr = 1/expr
    return l[0] + expr

syms = symbols('a0:4')
continued_frac = list_to_frac(syms)
continued_frac
```

## Symbolic Expressions

```python
# Symbolic expressions
expr = (a + b)**2
expr
```

```python
# expand
Eq(expr, expr.expand())
```

```python
# factor
expr = a**2 + 2*a*b + b**2
Eq(expr, expr.factor())
```

```python
# together
expr = 1/x + 1/y
Eq(expr, expr.together())
```

```python
# apart
expr = (x+y)/(x*y)
Eq(expr, expr.apart(x))
```

```python
# cancel
expr = (2*x**2 - 2)/(x**2 - 2*x + 1)
Eq(expr, expr.cancel())
```

```python
# simplify
expr = (2*x**2 - 2)/(x**2 - 2*x + 1)
Eq(expr, expr.simplify())
```
